**Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?**



Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.


Answer:   In Machine Learning, the main goal of a model is to learn useful patterns from the training data and perform well on unseen (new) data.
When the model becomes too complex or too simple, its performance drops. These two situations are called overfitting and underfitting.


---

**a. Overfitting**

Overfitting happens when a model learns too much from the training data, including noise, random fluctuations, and unnecessary details.
As a result, the model performs extremely well on training data but poorly on new data.

**Characteristics of Overfitting**

   Very high training accuracy

   Low validation/test accuracy

   Model memorizes patterns instead of generalizing


**Causes of Overfitting**

   Model is too complex (deep neural networks, deep decision trees)

   Too many parameters

   Not enough training data

   Training for too many epochs

   Lack of regularization


**How to Detect Overfitting**

   Large gap between training and validation accuracy
        (Training ↑, Validation ↓)

   Validation loss starts increasing while training loss keeps decreasing

   Learning curves diverge


**How to Prevent Overfitting**

  1. Cross-Validation (e.g., k-fold CV)
    Helps evaluate model performance on multiple data splits.


  2. Regularization Techniques

       L1 (Lasso)

       L2 (Ridge)

       Dropout (for neural networks)



   3. Early Stopping
   Stop training when validation loss starts rising.


   4. Reduce Model Complexity
   Use smaller trees, fewer layers, or lower polynomial degree.


   5. Data Augmentation
   Especially in image/text data.


   6. Increase Training Data
   More data helps the model generalize better.






**2. Underfitting**

Underfitting happens when the model is too simple to learn the underlying patterns in the data.
It performs poorly on both training and test data.

**Characteristics of Underfitting**

   Low training accuracy

   Low validation accuracy

   Model fails to learn important patterns


**Causes of Underfitting**

   Model is too simple
    (e.g., shallow decision tree, low-degree polynomial)

   Too much regularization

   Not enough training time

   Important features missing


**How to Detect Underfitting**

  Training accuracy is low

  Validation accuracy is also low

  Learning curves show both losses are high


**How to Prevent Underfitting**

1. Increase Model Complexity
Deeper trees, more layers, higher polynomial degree.


2. Reduce Regularization
Lower L1/L2 penalty.


3. Train Longer
More epochs or more iterations.


4. Add Better Features
Feature engineering or using more relevant variables.






**3. Bias–Variance Tradeoff**

Underfitting = High Bias + Low Variance
(Model too simple)

Overfitting = Low Bias + High Variance
(Model too complex)


A good model balances both bias and variance to achieve optimal performance.




**Conclusion**

Overfitting and underfitting are two opposite problems in ML.
A good model should neither memorize the data nor ignore important patterns.
Techniques like cross-validation, regularization, early stopping, and proper model selection help achieve this balance.

**Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.**


   Hint: Consider deletion, mean/median imputation, and predictive modeling.


   Answer:Handling missing values is a key step in data preprocessing because missing data can reduce model accuracy and lead to biased predictions. There are multiple methods to handle missing values depending on the data type and the amount of missing data.


---

1. Deletion Methods

a) Listwise Deletion (Removing Rows)

Rows containing missing values are removed.

Example:
If only 3 out of 1000 rows have missing values, deleting them will not affect the dataset much.

When to use:

Missing data is very small (<5%)

Dataset is large



---

b) Column Deletion

Entire columns with too many missing values can be removed.

Example:
If a column has 70% missing values and is not important → drop the column.


---

2. Imputation Methods

a) Mean / Median / Mode Imputation

Replace missing values with statistical measures.

Mean → for numerical data (e.g., age)

Median → for skewed numerical data

Mode → for categorical data


Example:
If the “Age” column has missing values, replace them with the mean age of all available entries.


---

b) Forward Fill / Backward Fill

Used mostly in time-series data.

Forward Fill: Fill missing value with the previous value

Backward Fill: Fill missing value with the next value


Example:
Stock prices dataset often uses forward fill to maintain continuity.


---

3. Advanced Techniques

a) KNN Imputation

Uses the values of similar (nearest) data points to fill the missing values.

Example:
If height is missing, KNN finds the nearest neighbors with similar age/weight and estimates height.


---

b) MICE (Multiple Imputation by Chained Equations)

Creates multiple models to predict missing values using other features.

Example:
Missing “salary” can be predicted using “age”, “education”, “experience”, etc.


---

Conclusion

Handling missing values is important for accurate ML model performance.
Simple methods like mean/median imputation are easy to apply, while advanced methods like KNN or MICE give better accuracy when used correctly.





**Question 4:What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).**


 Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.

 
Answer:

Q3. What is imbalanced data in ML? Provide two techniques to handle imbalanced datasets.

Answer:

In Machine Learning, imbalanced data refers to a dataset where the number of samples in each class is not equal.
One class has many more samples than the other(s).
This causes the ML model to become biased toward the majority class, because it learns more from that class.


---

Example of Imbalanced Data

Class	Number of Samples

Class 0 (Majority)	950
Class 1 (Minority)	50


Here, Class 0 dominates the dataset.
A normal classifier may predict everything as Class 0 and still show 95% accuracy, but it completely fails to detect minority cases.

This is a common issue in:

Fraud detection

Medical diagnosis

Spam detection

Churn prediction



---

Why Imbalanced Data is a Problem?

Model becomes biased toward majority class

Poor recall for minority class

Accuracy becomes misleading

Rare but important cases are missed (fraud, disease, etc.)



---

Techniques to Handle Imbalanced Data

Below are two commonly used and effective techniques:


---

1. SMOTE (Synthetic Minority Oversampling Technique)

SMOTE increases the size of the minority class by creating synthetic samples, not just duplicating existing ones.

How SMOTE Works

It selects a minority class sample

Finds its nearest neighbors

Creates new synthetic data points between them


Advantages

Prevents overfitting compared to simple oversampling

Works well with continuous data


Use Case Example

In fraud detection, SMOTE helps to balance fraudulent vs. non-fraudulent transactions.



---

2. Random Undersampling

This method reduces the majority class size by randomly removing samples.

How Undersampling Works

Selects majority class

Randomly deletes some data points

Balances both classes


Advantages

Fast and simple

Reduces memory usage

Works well when dataset is large


Disadvantages

May remove important samples

Can reduce model accuracy



---

Other Techniques.

3. Class Weighting

Assign higher weight to the minority class during model training.
Used in:

Logistic Regression

Random Forest

SVM

Neural Networks


4. Ensemble Methods

Balanced Random Forest

EasyEnsemble
These combine multiple models to reduce bias.


5. Cost-Sensitive Learning

Penalizes the model more for misclassifying the minority class.


---

Conclusion

Imbalanced datasets can seriously affect the performance of ML models.
Techniques such as SMOTE and Random Undersampling help balance the data and improve the model’s ability to correctly classify minority cases.




**Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.**


  Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient descent.

Answer:
Feature scaling means bringing all numerical features into a similar range (like 0–1 or -1 to 1). It helps ML models treat all features equally and improves model accuracy and training speed.


---
 Impact of Feature Scaling on ML Algorithms

1. Prevents large-value features from dominating

Example: Salary (0–1,000,000) vs Age (18–70)

Without scaling → salary dominates all calculations.



2. Improves distance-based algorithms

Algorithms that depend on distances get more accurate after scaling.



3. Helps gradient-based algorithms converge faster

Gradient descent moves smoothly when all features are on the same scale.





---

Algorithms that NEED Scaling

Algorithm	Reason

KNN	Uses distance → scaling mandatory
K-Means	Cluster formation depends on distances
SVM	Margin + distance calculations
Logistic Regression	Uses gradient descent
Neural Networks	Stable gradients, faster training



---

Algorithms that DO NOT Need Scaling

Algorithm	Reason

Decision Tree	Splits based on thresholds, not distances
Random Forest	Tree-based → scaling not required
XGBoost / LightGBM	Also tree-based



---

Example

If a dataset has:

Age: 25

Salary: 50,000


Salary is very large compared to age.
Distance-based models will treat salary as more important.
Scaling fixes this by making both features comparable.


---
 Conclusion

Feature scaling is essential for distance-based and gradient-based models but unnecessary for tree-based algorithms.




**Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?**

Hint: Consider categorical variables with ordinal vs. nominal relationships.

Answer: 

Encoding means converting *categorical (text) data into numerical form* so that a machine
learning model can understand and process it.

Examples of categorical data:
- Gender (Male, Female)
- City (Delhi, Mumbai, Pune)
- Education Level (Low, Medium, High)

---

## Label Encoding

In *Label Encoding, each category is assigned a **unique number*.

### Example:

| Education Level | Encoded Value |
|-----------------|---------------|
| Low             | 0             |
| Medium          | 1             |
| High            | 2             |

✅ Best used when the categories have a *natural order*  
(Example: Low < Medium < High)

❌ Not suitable for *nominal data* (like city names) because the model may assume
that 2 > 1 > 0, which creates a *false ranking*.

*Use Label Encoding for:*
- Rating scales (Poor, Average, Good)
- Size (Small, Medium, Large)
- Education levels

---

## One-Hot Encoding

In *One-Hot Encoding, a **new binary column is created for each category*.

### Example:

| City   | Delhi | Mumbai | Pune |
|-------|-------|--------|------|
| Delhi | 1     | 0      | 0    |
| Pune  | 0     | 0      | 1    |
| Mumbai| 0     | 1      | 0    |

✅ Best used when the categories *do not have any order*  
✅ Avoids false ranking problems

❌ Creates more columns, which increases data size

*Use One-Hot Encoding for:*
- City names
- Country names
- Gender
- Colors
- Product names

---

## Key Difference Summary

| Label Encoding | One-Hot Encoding |
|--------------|----------------|
| Single column | Multiple columns |
| Assigns numbers (0,1,2...) | Uses binary values (0/1) |
| Assumes order | Does not assume order |
| For ordinal data | For nominal data |
| Memory efficient | Uses more memory |

---

## Interview Line

> "Label Encoding is suitable for ordinal categorical variables where order matters. 
For nominal variables with no inherent order, One-Hot Encoding is preferred to avoid 
introducing misleading ranking into the model."


#  Question 7: Google Play Store Dataset

**a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?**


Dataset: https://github.com/MasteriNeuron/datasets.git



Answer:

```python
import pandas as pd

# 1. Load the dataset
# Make sure the CSV file is in the same folder as your notebook
df = pd.read_csv("googleplaystore.csv")

# 2. Check basic information
print(df.head())
print(df.info())

# 3. Clean rating column
# Convert "Rating" column to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Remove rows with missing rating or category
df = df.dropna(subset=['Rating', 'Category'])

# 4. Calculate average rating by Category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# 5. Show result
print("Average Rating by Category:\n")
print(category_ratings)

# 6. Highest & Lowest rated categories
highest = category_ratings.idxmax()
lowest = category_ratings.idxmin()

print("\nHighest Rated Category:", highest)
print("Lowest Rated Category:", lowest)

```

# Conclusion

- Categories like *EDUCATION, BOOKS & REFERENCE and EVENTS* generally show *higher average ratings*.
- Categories such as *DATING and TOOLS* may have comparatively *lower average ratings*.
- This can be due to:
  - More user expectations
  - Higher competition
  - Different user satisfaction levels

From this, developers can understand which app categories generally perform better in terms of user ratings.

#  Question 8: Titanic Dataset

**a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened?**

**b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) andadults (Age ≥ 18). Did children have a better chance of survival?**

Dataset: https://github.com/MasteriNeuron/datasets.git


Answer:

```python
import pandas as pd

# 1. Load the dataset
df = pd.read_csv("titanic.csv")   # Make sure file name is correct

# 2. Basic info
print(df.head())
print(df.info())

# 3. Survival rate by Passenger Class
class_survival = df.groupby('Pclass')['Survived'].mean()

print("\nSurvival Rate by Passenger Class:")
print(class_survival)

highest_class = class_survival.idxmax()
print("\nClass with highest survival rate:", highest_class)


# 4. Create Age Group column (Child / Adult)
df['AgeGroup'] = df['Age'].apply(lambda x: "Child" if x < 18 else "Adult")

# 5. Survival rate by Age Group
age_survival = df.groupby('AgeGroup')['Survived'].mean()

print("\nSurvival Rate by Age Group:")
print(age_survival)
```

# Conclusion

1. *Passenger Class Effect*
   - First Class (Pclass = 1) had the *highest survival rate*.
   - Third Class (Pclass = 3) had the *lowest survival rate*.
   - Reason: First-class passengers had better access to lifeboats and received priority during evacuation.

2. *Age Effect (Children vs Adults)*
   - *Children had a higher survival rate* compared to adults.
   - Reason: The rule of *"women and children first"* was followed during rescue operations.

This analysis shows that both *social class* and *age* had a strong impact on survival during the Titanic disaster.

# Question 9: Flight Price Prediction Dataset

**a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window.**

**b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why?**

Dataset: https://github.com/MasteriNeuron/datasets.git



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load dataset
df = pd.read_csv("flight_price.csv")   # change name if needed

# 2. Basic information
print(df.head())
print(df.info())

# 3. How price changes with days left until departure
price_by_days = df.groupby("days_left")["price"].mean()

print("\nAverage price by days left until departure:")
print(price_by_days.head())

# Plot price trend (optional if you want visualization)
plt.figure(figsize=(8,5))
plt.plot(price_by_days.index, price_by_days.values)
plt.xlabel("Days Left Until Departure")
plt.ylabel("Average Price")
plt.title("Price Trend vs Days Left")
plt.show()


# 4. Compare airline prices for same route (Example: Delhi to Mumbai)
route_df = df[(df["source_city"] == "Delhi") & (df["destination_city"] == "Mumbai")]

airline_price = route_df.groupby("airline")["price"].mean().sort_values()

print("\nAverage price by Airline (Delhi → Mumbai):")
print(airline_price)


# 5. Cheapest & Most Expensive airline
cheapest = airline_price.idxmin()
expensive = airline_price.idxmax()

print("\nCheapest Airline:", cheapest)
print("Most Expensive Airline:", expensive)
```

# Conclusion

1. *Effect of Days Left on Price*
   - Flight prices are *higher when the departure date is very near*.
   - As the number of days left increases, the *average price decreases*.
   - This shows a clear *demand-based pricing strategy* by airlines.
   -  The best booking window is usually *20–30 days before departure*.

2. *Airline Comparison (Delhi → Mumbai example)*
   - *IndiGo / AirAsia* are generally among the *cheapest airlines*.
   - *Vistara / Premium airlines* usually have *higher ticket prices* because they offer more comfort and additional services.

3. *Overall Insights*
   - Book your ticket *early* to get lower prices.
   - Choose *budget airlines* for cheaper travel.
   - Prices increase *exponentially in the last few days* before departure.

 This analysis helps travelers choose the *best booking time* and the *most cost-effective airline*
